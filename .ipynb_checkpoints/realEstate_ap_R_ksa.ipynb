```python
import ast
def extract_date(cell):
    # Convert the string representation to a list
    parsed_list = ast.literal_eval(cell)

    # Extract the date and time from the list
    date_str = parsed_list[0]  # Assuming the first index is the date
    time_str = parsed_list[1]  # Assuming the second index is the time

    # Combine date and time strings
    datetime_str = f"{date_str} {time_str}"

    # Convert the combined string to a datetime object
    return pd.to_datetime(datetime_str)
```

```python
apdf['last_update'] = apdf['last_update'].apply(extract_date)
apdf['create_time'] = apdf['create_time'].apply(extract_date)
```

```python
apdf['last_update']=apdf['last_update'].astype('datetime64[ns]')
apdf['create_time']=apdf['create_time'].astype('datetime64[ns]')
```
